## So, We are going to train the model using California housing Price 1980 dataset that model can predict the house price in the bases of features. and the r2 score of this model will going to be 82.12.

#### Importing Libraries-

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import r2_score
import pickle
```

##### Loading the csv dataset-

```python
df = pd.read_csv("housing.csv")
```

##### Checking null values of dataset-

```python
df.isnull().sum()
```

##### Checking datatypes of columns-

```python
df.info()
```

##### Filling NaN values-

```python
df["total_bedrooms"] = df["total_bedrooms"].fillna(df["total_bedrooms"].median())
```

##### Statical infomation about dataset-

```python
df.describe()
```

##### Visualization of data-

```python
sns.histplot(df["housing_median_age"], bins=50)
plt.show()
```

```python
sns.histplot(df["median_house_value"], bins=50)
plt.show()
```

##### Removing outliers-

```python
df.drop(df[df["median_house_value"] > 495000].index, inplace=True)
```

```python
df = df[df["housing_median_age"] < 51]
```

##### Seprating numerical and Categorial columns name for ColumnTransformer-

```python
cat_col = ["ocean_proximity"]
num_col = ["longitude", "latitude", "housing_median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income"]
```

##### building pipeline for each datatype columns-

```python
num_pipe = Pipeline([
    ("scale", StandardScaler())
])
cat_pipe = Pipeline([
    ("encode", OneHotEncoder(handle_unknown="ignore"))
])
```

##### Making full Pipeline that will apply pipelines for given columns names-

```python
full_pipe = ColumnTransformer([
    ("num", num_pipe, num_col),
    ("cat", cat_pipe, cat_col)
])
```

##### Seprating Features(X) and label(y) for model Training-

```python
X = full_pipe.fit_transform(df[["longitude", "latitude", "housing_median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income", "ocean_proximity"]])

y = df["median_house_value"].values    
```

##### Spiliting the data for Model traing and testing-

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

##### Initializing and fitting the data to best model with hyperparameters for this dataset-

```python
gbs_reg = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.03, max_depth=8, min_samples_split=8, min_samples_leaf=2, subsample=0.8, max_features=500, random_state=2)
gbs_reg.fit(X_train, y_train)
```

##### Initializing the prediction of the model-

```python
gbs_pred = gbs_reg.predict(X_test)
```

##### Checking the r2_score of the Model-

```python
r2 = r2_score(y_test, gbs_pred)
print(r2)
```

##### Checking cross value score of model-

```python
gbs_rmses = -cross_val_score(gbs_reg, X_train, y_train, scoring="neg_root_mean_squared_error", cv=10)
print("gbs", pd.Series(gbs_rmses).describe())
```

##### Visualizing Prediction in graph-

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, gbs_pred, alpha=0.5, color='blue', label='Predicted vs Actual')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Scatter Plot: Predicted vs Actual')
plt.legend()
plt.grid(True)
plt.show()
```
